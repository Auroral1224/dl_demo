## Download pretrained weights
Run this once to grab all YOLOv5/v6/v7 checkpoints into `weights/`.

````python
%%bash
set -euo pipefail
mkdir -p weights
download() {
  url=$1
  file=$2
  if [ -f "weights/$file" ]; then
    echo "weights/$file already exists, skipping."
  else
    wget -O "weights/$file" "$url"
  fi
}
download https://github.com/ultralytics/yolov5/releases/download/v7.0/yolov5s.pt yolov5s.pt
download https://github.com/ultralytics/yolov5/releases/download/v7.0/yolov5m.pt yolov5m.pt
download https://github.com/ultralytics/yolov5/releases/download/v7.0/yolov5l.pt yolov5l.pt
download https://github.com/ultralytics/yolov5/releases/download/v7.0/yolov5x.pt yolov5x.pt
download https://github.com/meituan/YOLOv6/releases/download/0.4.0/yolov6n.pt yolov6n.pt
download https://github.com/meituan/YOLOv6/releases/download/0.4.0/yolov6s.pt yolov6s.pt
download https://github.com/meituan/YOLOv6/releases/download/0.4.0/yolov6m.pt yolov6m.pt
download https://github.com/meituan/YOLOv6/releases/download/0.4.0/yolov6l.pt yolov6l.pt
download https://github.com/WongKinYiu/yolov7/releases/download/v0.1/yolov7.pt yolov7.pt
download https://github.com/WongKinYiu/yolov7/releases/download/v0.1/yolov7x.pt yolov7x.pt
````
